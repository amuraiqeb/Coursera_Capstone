## 1.1. Business Problem

A young entrepreneur has decided to open a coffee shop in London. The project is financially viable, but the location of the shop has not been selected yet. To make the best decision, the entrepreneur needs to know which borough in London can be the most profitable. We will need to explore the venues in London's boroughs to understand where the competition is located. This will help in determining the best location for the new shop. Also, we need to know which customers these competing venues serve. To this end, we will explore the boroughs further to predict how the new shop can be supported by nearby venues such as university campuses, office spaces, and other cultural venues. Ideally, the new coffee shop will be located in the borough with the highest concentration of potential customers and the lowest competing venues' concentration.
