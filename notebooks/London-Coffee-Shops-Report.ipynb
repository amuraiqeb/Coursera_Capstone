# 1. Introduction

## 1.1. Business Problem

A young entrepreneur has decided to open a coffee shop in London. The project is financially viable, but the location of the shop has not been selected yet. To make the best decision, the entrepreneur needs to know which borough in London can be the most profitable. We will need to explore the venues in London's boroughs to understand where the competition is located. This will help in determining the best location for the new shop. Also, we need to know which customers these competing venues serve. To this end, we will explore the boroughs further to predict how the new shop can be supported by nearby venues such as university campuses, office spaces, and other cultural venues. Ideally, the new coffee shop will be located in the borough with the highest concentration of potential customers and the lowest competing venues' concentration.

## 1.2. Data Sources

The data required for this project can be collected from multiple sources. The project relies primarily on the foursqure API to collect data about the venues in London and their location. The Greater London Area will be covered in this project. The data is publicly available and is collected and organised from the wikipedia <code>List of areas of London</code> page which can be found [HERE](https://en.wikipedia.org/wiki/List_of_areas_of_London). The data contains information about the location of the neighborhoods, their boroughs, towns, postcodes, dial codes, and OS grid ref. This project utilizes the columns <code>Location</code>, <code>London borough</code>, <code>Post town</code>, <code>Postcode district</code>; the other columns are not required and droped from the dataframe. The data also requires cleaning, especially the data related to boroughs as they contain unecessary references. The <code>Postcode district</code> column contains cells that have multiple postcodes that have been split into two rows to standarize the data and prepare it for geocoding. The data is then complemented with Latitude and Longtitude data for the different postcodes which are collected using <code>geocoder</code> and <code>ArcGIS</code>.
